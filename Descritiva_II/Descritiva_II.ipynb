```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('sinasc_RO_2019.csv')
df.head(10)
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
sns.set(style="whitegrid")

# Criando um boxplot da idade da mãe
plt.figure(figsize=(10, 6))
sns.boxplot(x='IDADEMAE', data=df, color='skyblue')
plt.title('Boxplot da Idade da Mãe')
plt.xlabel('Idade da Mãe')
plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
sns.set(style="whitegrid")

# Criando um boxplot do peso do bebê
plt.figure(figsize=(10, 6))
sns.boxplot(x='PESO', data=df, color='salmon')
plt.title('Boxplot do Peso do Bebê')
plt.xlabel('Peso do Bebê (g)')
plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
sns.set(style="whitegrid")

# Criando um histograma da idade da mãe
plt.figure(figsize=(10, 6))
sns.histplot(df['IDADEMAE'], bins=30, kde=True, color='skyblue')
plt.title('Histograma da Idade da Mãe')
plt.xlabel('Idade da Mãe')
plt.ylabel('Contagem')
plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
sns.set(style="whitegrid")

# Criando um histograma do peso do bebê
plt.figure(figsize=(10, 6))
sns.histplot(df['PESO'], bins=30, kde=True, color='salmon')
plt.title('Histograma do Peso do Bebê')
plt.xlabel('Peso do Bebê (g)')
plt.ylabel('Contagem')
plt.show()
```
