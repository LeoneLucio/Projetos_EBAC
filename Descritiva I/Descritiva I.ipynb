#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('sinasc_RO_2019.csv')
df.head(10)

```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
plt.figure(figsize=(12, 8))
pd.crosstab(df['APGAR1'], df['SEXO']).plot(kind='bar', stacked=True, color=['pink', 'lightblue'])
plt.title('Relação entre APGAR1 e SEXO')
plt.xlabel('APGAR1')
plt.ylabel('Contagem')
plt.show()

# Gráfico de barras empilhadas para a relação entre GESTACAO e ESTCIVMAE
plt.figure(figsize=(12, 8))
pd.crosstab(df['GESTACAO'], df['ESTCIVMAE']).plot(kind='bar', stacked=True, color=['lightgreen', 'lightcoral', 'lightblue'])
plt.title('Relação entre GESTACAO e ESTCIVMAE')
plt.xlabel('GESTACAO')
plt.ylabel('Contagem')
plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
avg_apgar5_by_consultas = df.groupby('CONSULTAS')['APGAR5'].mean()

# Criando o gráfico de barras
plt.figure(figsize=(12, 8))
avg_apgar5_by_consultas.plot(kind='bar', color='lightblue')
plt.title('Média do APGAR5 por quantidade de CONSULTAS')
plt.xlabel('Quantidade de CONSULTAS')
plt.ylabel('Média do APGAR5')
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
sns.set(style="whitegrid")

# Criando um painel com 2 linhas e 2 colunas
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(14, 10))

# Gráfico de barras para SEXO
sns.countplot(x='SEXO', data=df, palette='pastel', ax=axes[0, 0])
axes[0, 0].set_title('Distribuição de SEXO')

# Gráfico de barras para GESTACAO
sns.countplot(x='GESTACAO', data=df, palette='pastel', ax=axes[0, 1])
axes[0, 1].set_title('Distribuição de GESTACAO')

# Gráfico de barras para ESTCIVMAE
sns.countplot(x='ESTCIVMAE', data=df, palette='pastel', ax=axes[1, 0])
axes[1, 0].set_title('Distribuição de ESTCIVMAE')

# Gráfico de barras para CONSULTAS
sns.countplot(x='CONSULTAS', data=df, palette='pastel', ax=axes[1, 1])
axes[1, 1].set_title('Distribuição de CONSULTAS')

# Ajustando o layout
plt.tight_layout()

# Mostrando os gráficos
plt.show()
```
